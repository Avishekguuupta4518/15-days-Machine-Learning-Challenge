## What is Machine Learning?

Machine Learning (ML) is a subset of Artificial Intelligence that allows systems
to learn patterns from data and improve their performance without being explicitly programmed.

### Intuition:
- Traditional Programming: Rules + Data → Output
- Machine Learning: Data + Output → Rules (Model)

ML is useful when:
- Data is large
- Patterns are complex
- Rules are hard to define manually


## ML vs AI vs Deep Learning

### Artificial Intelligence (AI)
- Broad field
- Any system that mimics human intelligence
- Includes rule-based systems and ML

### Machine Learning (ML)
- Subset of AI
- Learns from data
- Examples: regression, classification

### Deep Learning (DL)
- Subset of ML
- Uses neural networks with many layers
- Examples: image recognition, speech recognition

### Relationship:
AI ⟶ ML ⟶ Deep Learning


## Types of Machine Learning

### 1. Supervised Learning
- Data has labels
- Used for prediction
- Examples: house price prediction, spam detection

### 2. Unsupervised Learning
- No labels
- Finds hidden patterns
- Examples: customer segmentation, clustering

### 3. Reinforcement Learning
- Learns by trial and error
- Uses rewards and penalties
- Examples: game AI, robotics
